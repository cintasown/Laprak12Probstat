```python
# Muhammad Fadhil Ramadhaa - 064102400010
import numpy as np
import matplotlib.pyplot as plt

# Data
ukuran = np.array([40, 55, 60, 70, 85, 100, 120, 150, 200, 250])
harga = np.array([1.8, 4.2, 4.5, 5.2, 6.0, 7.2, 8.4, 10.5, 14.2, 17.0])

# Scatter plot
plt.scatter(ukuran, harga, color='darkgreen')
plt.title("Ukuran Apartemen vs Harga Sewa")
plt.xlabel("Ukuran (m²)")
plt.ylabel("Harga Sewa (juta/bulan)")
plt.grid(True)
plt.show()
```

```python
# Muhammad Fadhil Ramadhan - 064102400010
from sklearn.linear_model import LinearRegression

# Ubah ukuran jadi matriks kolom
X = ukuran.reshape(-1, 1)

# Buat model regresi linier
model = LinearRegression()
model.fit(X, harga)

# Prediksi harga dengan model regresi
prediksi_harga = model.predict(X)
```

```python
# Muhammad Fadhil Ramadhan - 064102400010
import matplotlib.pyplot as plt

plt.scatter(ukuran, harga, label="Data Asli", color='darkgreen')
plt.plot(ukuran, prediksi_harga, color='orange', label="Garis Regresi")
plt.xlabel("Ukuran (m²)")
plt.ylabel("Harga Sewa (juta/bulan)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Muhammad Fadhil Ramadhan - 064102400010
import numpy as np
corr = np.corrcoef(ukuran, harga)[0,1]
print(f"Korelasi Pearson : {corr:.3f}")
```

```python
#Muhammad Fadhil Ramadhan - 064102400010
from sklearn.metrics import r2_score
r2=r2_score(harga, prediksi_harga)
print(f"Koefisien Determinasi (R^2): {r2:.3f}")
```

```python
#Muhammad Fadhil Ramadhan - 064102400010
x_baru = np.array([[175]]) 
y_baru = model.predict(x_baru)
print(f"Prediksi harga sewa untuk 175 m^2: {y_baru[0]:.2f} juta/bulan")
```
